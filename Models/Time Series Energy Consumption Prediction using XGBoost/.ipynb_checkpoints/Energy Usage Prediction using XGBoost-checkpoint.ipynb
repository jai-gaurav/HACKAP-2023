### Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from sklearn.metrics import mean_squared_error

plt.style.use('fivethirtyeight')

%matplotlib inline
```

### Importing dataset

```python
df = pd.read_csv("PJME_hourly.csv")
df = df.set_index("Datetime")
df.index = pd.to_datetime(df.index)
df.head()
```

### Draw basic timeseries plot

```python
color_pal = sns.color_palette()

df.plot(style='.', 
        figsize=(15,5), 
        color=color_pal[0], 
        title="PJME Energy Use (MW)")
```

### Train test split and plot

```python
train = df.loc[df.index < '01-01-2015']
test = df.loc[df.index >= '01-01-2015']

fig, ax = plt.subplots(figsize=(15,5))
train.plot(ax=ax, label="Training Dataset")
test.plot(ax=ax, label="Testing Dataset")
ax.axvline('01-01-2015', color='black', ls='--')
ax.legend(['Training Dataset', 'Testing Dataset'])
plt.show()
```

### See weekly distribution for patterns

```python
week = df.loc[(df.index > '01-01-2010') & (df.index < '01-08-2010')]
week.plot(figsize=(15,5), title="Weekly Distribution")
plt.show()
```

### Feature creation

```python
def createFeatures(df):
    df = df.copy()
    df['hour'] = df.index.hour
    df['dayofweek'] = df.index.dayofweek
    df['quarter'] = df.index.quarter
    df['month'] = df.index.month
    df['year'] = df.index.year
    df['dayofyear'] = df.index.dayofyear
    return df

df = createFeatures(df)
```

### Visualize feature target relationship

```python
# hour - MW relationship
fig, ax = plt.subplots(figsize=(10,8))
sns.boxplot(data=df, x='hour', y='PJME_MW')
ax.set_title("MW by hour")
```

```python
# month - MW relationship
fig, ax = plt.subplots(figsize=(10,8))
sns.boxplot(data=df, x='month', y='PJME_MW')
ax.set_title("MW by month")
```

```python
# year - MW relationship
fig, ax = plt.subplots(figsize=(10,8))
sns.boxplot(data=df, x='year', y='PJME_MW')
ax.set_title("MW by year")
```

```python
# dayofweek - MW relationship
fig, ax = plt.subplots(figsize=(10,8))
sns.boxplot(data=df, x='dayofweek', y='PJME_MW')
ax.set_title("MW by dayofweek")
```

```python
# dayofyear - MW relationship
fig, ax = plt.subplots(figsize=(10,8))
sns.boxplot(data=df, x='dayofyear', y='PJME_MW')
ax.set_title("MW by dayfyear")
```

```python
# quarter - MW relationship
fig, ax = plt.subplots(figsize=(10,8))
sns.boxplot(data=df, x='quarter', y='PJME_MW')
ax.set_title("MW by quarter")
```

### Creating Model

```python
# adding features in both training and testing datasets
train = createFeatures(train)
test = createFeatures(test)

# choosing features and target
FEATURES = ['hour', 'dayofweek', 'quarter', 'month', 'year', 'dayofyear']
TARGET = 'PJME_MW'
```

```python
X_train = train[FEATURES]
Y_train = train[TARGET]

X_test = test[FEATURES]
Y_test = test[TARGET]
```

```python
reg = xgb.XGBRegressor(n_estimators = 1000, 
                       early_stopping_rounds=50,
                      learning_rate = 0.01)
reg.fit(X_train, Y_train, 
        eval_set = [(X_train, Y_train), (X_test, Y_test)],
       verbose=100)
```

### Feature Importance

```python
fi = pd.DataFrame(data = reg.feature_importances_,
             index = reg.feature_names_in_,
             columns = ['importance'])
reg.feature_importances_
```

```python
fi.sort_values('importance', ascending=False).plot(kind='bar', title='Feature Importance')
plt.show()
```

### Forecast on test dataset

```python
test['prediction'] = reg.predict(X_test)
df_final = df.merge(test[['prediction']], 
                    how='left', 
                    left_index=True, 
                    right_index=True)
df_final = df_final.loc[df_final.index >= '01-01-2015']
```

```python
ax = df_final[['PJME_MW']].plot(figsize=(15,5))
df_final['prediction'].plot(ax=ax, style='.')
ax.set_title("Raw Data and Predictions")
ax.legend(['True Data', 'Predictions'])
plt.show()
```

```python
week = df_final.loc[(df_final.index > '01-01-2015') & (df_final.index < '01-08-2015')]
week = week[['PJME_MW', 'prediction']]
week.plot(figsize=(15,5), title="Weekly Distribution")
plt.legend(['True Data', 'Predictions'])
plt.show()
```

### Using test metric

```python
score = np.sqrt(mean_squared_error(test['PJME_MW'], test['prediction']))
print(f"RMSE Score on the test set: {score:0.2f}")
```

### Calculate Error

```python
# worst 5 days
test['error'] = np.abs(test[TARGET] - test['prediction'])
test['date'] = test.index.date
test.groupby(['date'])['error'].mean().sort_values(ascending=False).head(5)
```

```python
# best 5 days
test['error'] = np.abs(test[TARGET] - test['prediction'])
test['date'] = test.index.date
test.groupby(['date'])['error'].mean().sort_values().head(5)
```
